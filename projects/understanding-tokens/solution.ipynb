Tokenizing with code

```python
import tiktoken
```

```python
encoding = tiktoken.encoding_for_model("gpt-4.1-mini")
tokens=encoding.encode("Hello, world!")
print(tokens)
```

```python
for token_id in tokens:
    print(f"{token_id}: {encoding.decode([token_id])}")
```
